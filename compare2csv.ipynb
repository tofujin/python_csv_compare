```python
import pandas as pd
import os 
import os.path
import numpy as np
import pandas.io.sql as psql
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

# Given Input

```python
# read two input files with delimieter common (",")
ref=pd.read_csv('70210444b.csv', sep=",",low_memory=False)
new=pd.read_csv('70210464b.csv', sep=",",low_memory=False)

compare_key ='compare_key'
compare_field = ["compare_field1","compare_field2"]

# convert null to ''
ref = ref.fillna('')
new = new.fillna('')

ref_header_array = ref.columns.values
new_header_array = new.columns.values

# replace the header with "ref" in ref file
for x in ref_header_array:
    ref.rename(columns={x:x+'_ref'}, inplace=True)

for y in new_header_array:
    new.rename(columns={y:y+'_new'}, inplace=True)

```

```python
print("No. of records in reference run = ", len(ref))
#print("10 Sample Records")
ref_sample = ref.sample(n = len(ref))
ref_sample

#lpc.as_of_date.value_counts()
```

```python
print("No. of records in new run = ", len(new))
#new_sample = new.sample(n = len(new))
#new_sample

#lpc.as_of_date.value_counts()
```

```python
input_join = pd.merge(new,ref,how='inner',right_on=[compare_key+'_ref'], left_on=[compare_key+'_new'])
print("NUMBER OF RECORDS mapped ", len(input_join))
input_join_sample = input_join.sample (n=len(input_join))
input_join_sample
```

```python
compare_condition = ""
compare_len = len(ref_header_array)

for i in range(1,compare_len):
    if len(compare_condition) == 0:
        if ref_header_array[i] in compare_field:
            compare_condition = compare_condition + "(input_join['"+ref_header_array[i]+"_ref'] != input_join['" + ref_header_array[i]+"_new'])"
    else:
        if ref_header_array[i] in compare_field:
            compare_condition = compare_condition + ' | '
            compare_condition = compare_condition + "(input_join['"+ref_header_array[i]+"_ref'] != input_join['" + ref_header_array[i]+"_new'])"

print(compare_condition)

input_join_mismatch = input_join[eval(compare_condition)]


print("No. of records mistmatch = ", len(input_join_mismatch))
input_join_mismatch
```

## Output mismatch result to file

```python
writer = pd.ExcelWriter('compare_result.xlsx', engine='xlsxwriter')
input_join_mismatch.to_excel(writer, sheet_name='DIFF', index=False)
writer.save()
```
